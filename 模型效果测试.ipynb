```python
import matplotlib.pyplot as plt
```

```python
import json

y1 = json.load(open("log/cross_entropy_acc.json"))
y2 = json.load(open("log/fake_cross_entropy_acc.json"))
x = range(1, len(y1) + 1)
```

```python
plt.figure(figsize = (15, 7))
plt.plot(x, y1, marker = 'o', label = "cross entropy")
plt.plot(x, y2, marker = 'x', label = "fake cross entropy")
plt.legend()
plt.show()
```
